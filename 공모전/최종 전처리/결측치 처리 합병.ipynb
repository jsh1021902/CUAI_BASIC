```python
import pandas as pd
```

```python
df1 = pd.read_csv("C:/Users/MSI/Desktop/CUAI/Basic/우주선 타이타닉_베이직/결측치 처리_05.18/Home,Cryo_성현우")
df2 = pd.read_csv("C:/Users/MSI/Desktop/CUAI/Basic/우주선 타이타닉_베이직/결측치 처리_05.18/Cabin,Dest_김예원")
df3 = pd.read_csv("C:/Users/MSI/Desktop/CUAI/Basic/우주선 타이타닉_베이직/결측치 처리_05.18/Age,VIP_정서현")
df4 = pd.read_csv("C:/Users/MSI/Desktop/CUAI/Basic/우주선 타이타닉_베이직/결측치 처리_05.18/Spend_김지호")
```

```python
df = pd.concat([df1, df2, df3, df4],axis=1)
```

```python
df
```

```python
df.drop(columns = ['Unnamed: 0'], inplace=True)
```

```python
df = df[['PassengerId', 'HomePlanet', 'CryoSleep', 'Cabin_side', 
         'Cabin_deck', 'Cabin_number', 'Destination', 'Age', 
         'Age_group', 'VIP', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Transported']]
```

```python
df
```

```python
one_hot_df1 = pd.get_dummies(df, columns=["HomePlanet","CryoSleep","Destination","Cabin_deck","Cabin_side", 'VIP'], drop_first=True)
one_hot_df1["Transported"].replace({True:1,False:0}, inplace=True)
```

```python
one_hot_df1
```

```python
one_hot_df2 = pd.get_dummies(df, columns=["HomePlanet","CryoSleep","Destination","Cabin_deck","Cabin_side", "Age_group", 'VIP'], drop_first=True)
one_hot_df2["Transported"].replace({True:1,False:0}, inplace=True)
```
